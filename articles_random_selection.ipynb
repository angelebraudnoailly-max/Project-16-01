### Random selection of articles in "CAIRN_real_final_chunks.csv" based on a random selection of authors.


### 1. Library Import

```python
import pandas as pd
```

### 2. Data Import 

```python
df1=pd.read_csv('CAIRN_real_final_chunks.csv')
display (df1.head())
```

```python
df2=pd.read_csv('LM_real_final_chunks.csv')
display (df2.head())
```

### 3. Random Section of authors 

#### 3.a In CAIRN

```python
sample_values = df1['name'].sample(n=4, random_state=None)

print(sample_values.tolist())
```

#### 3.b In Le Monde

```python
sample_values2 = df2['russian_author_names'].sample(n=4, random_state=None)

print(sample_values2.tolist())
```

### 4. Random Selection of articles speaking of the chosen authors

#### 4.a In CAIRN

```python
selected_rows = {}

for author in sample_values:
    # Filter rows matching this author
    author_rows = df1[df1['name'] == author]
    
    # Randomly select 4 rows for this author (or fewer if not enough rows available)
    selected_rows[author] = author_rows.sample(n=min(3, len(author_rows)), random_state=None)

# Concatenate all selected rows into a final DataFrame
final_selection = pd.concat(selected_rows.values())

display(final_selection)

final_selection.to_csv('sample_CAIRN_chunk.csv', index=False)
```

#### 4.2 In Le Monde

```python
selected_rows2 = {}

for author2 in sample_values2:
    # Filter rows matching this author
    author_rows2 = df2[df2['russian_author_names'] == author2]
    
    # Randomly select 4 rows for this author (or fewer if not enough rows available)
    selected_rows2[author2] = author_rows2.sample(n=min(2, len(author_rows2)), random_state=None)

# Concatenate all selected rows into a final DataFrame
final_selection2 = pd.concat(selected_rows2.values())

display(final_selection2)

final_selection2.to_csv('sample_LM_chunk.csv', index=False)
```

### 5. Downloading in Excel format

```python

final_selection.to_excel('sample_CAIRN_chunk.xlsx', index=False) #It facilitates the reading of the file for me
```

```python
final_selection2.to_excel('sample_LM_chunk.xlsx', index=False) 
```
