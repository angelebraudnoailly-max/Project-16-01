```python
from openai import OpenAI
import pandas as pd
import re
```

### 2.Data Import & Connexion to LM studio

```python
#Import of CAIRN dataset 
df_textes = pd.read_csv("sample_CAIRN_chunk.csv")
display(df_textes.head())

# Initialize the OpenAI client with a custom base URL
client = OpenAI(base_url="http://localhost:1234/v1", api_key="not-needed")
```

```python
#Import of Le Monde dataset
df_textes2 = pd.read_csv("sample_LM_chunk.csv")
display(df_textes2.head())
```

### 3. LLM Data Analysis, with llama-3-8b

#### 3.a CAIRN

```python
df_textes["Centralité de la littérature"] = ""
df_textes["Noms_propres_identifies"] = ""
df_textes["Contexte_citation_auteur_russe"] = ""

for i, row in df_textes.iterrows():
    try:
        texte = row["chunk"]
        auteur_russe_cible = row["variante_recherchee"]
        
        prompt = f"""ANALYSE DE TEXTE - RÉPONDS UNIQUEMENT DANS LE FORMAT EXACT CI-DESSOUS

TEXTE À ANALYSER :
{texte}

AUTEUR RUSSE CIBLE À ANALYSER : {auteur_russe_cible}

---

INSTRUCTIONS :

1. CENTRALITÉ DE LA LITTÉRATURE :
• 0 = Littérature ABSENTE : le texte n'aborde pas du tout le thème de la littérature. Les auteurs sont mentionnés uniquement comme symboles culturels, historiques ou politiques, sans aucun lien avec leurs œuvres littéraires.
• 1 = Littérature CENTRALE : le texte traite principalement et explicitement de littérature (analyse d'œuvres, critique littéraire, théorie de l'écriture, esthétique littéraire, rôle social de la littérature).
• 2 = Littérature ÉVOQUÉE : le texte ne parle pas principalement de littérature, mais la mention d'auteurs sert à introduire ou illustrer le thème littéraire de manière secondaire.

2. IDENTIFICATION DES NOMS PROPRES :
• Si le texte contient des noms propres de personnes, liste-les avec leur nombre d'apparitions RÉEL.
• Format : Nom(nombre/nationalité/auteur_littéraire_oui_non)
• COMPTE VRAIMENT les occurrences : si un nom apparaît 1 fois → "1", 2 fois → "2", etc.
• Si possible, associe la nationalité à un pays.
• Si AUCUN nom propre n'est présent dans le texte, écrire SEULEMENT "aucun" sans autres mots.

3. CONTEXTE DE CITATION DE L'AUTEUR RUSSE CIBLE :
• "absent" = l'auteur russe cible n'est pas mentionné dans le texte
• "seul" = l'auteur russe cible est mentionné isolément, sans autres auteurs dans la même phrase
• "parmi_d'autres" = l'auteur russe cible est cité dans une énumération avec d'autres auteurs dans la même phrase

EXEMPLE CORRECT :
Texte n1: "Dostoïevski est un grand écrivain. La théorie du livre 'La Religieuse' (Diderot, n1, 1780) est aussi liée à celle de Tolstoï et de Dostoïevski."
Réponse n1: NOMS_PROPRES: Dostoïevski(2/Russe/oui)/Diderot(1/Français/oui)/Tolstoï(1/Russe/oui)
Texte n2 : "Il fut un grand empereur et a marqué l'histoire de France."
Réponse n2 : NOMS_PROPRES: aucun
---

RÉPONSE ATTENDUE (FORMAT EXACT) :

CENTRALITE:[0|1|2]
NOMS_PROPRES:[Nom1(nombre/nationalité/auteur_littéraire_oui_non)/Nom2(nombre/nationalité/auteur_littéraire_oui_non)/...] OU [aucun]
CONTEXTE_AUTEUR_RUSSE:[absent|seul|parmi_d'autres]

---

COMMENCE TA RÉPONSE DIRECTEMENT PAR "CENTRALITE:" :"""

        response = client.chat.completions.create(
            model="llama-3-8b-instruct-32k-v0.1",
            messages=[{"role": "user", "content": prompt}],
            temperature=0,
            max_tokens=500
        )
        
        answer = response.choices[0].message.content.strip()
        print(f"Réponse ligne {i+1} : {answer}")
        
        # Data extraction
        centralite_match = re.search(r'CENTRALIT[ÉE]:?\s*(\d)', answer)
        noms_match = re.search(r'NOMS_PROPRES:([^\n]+)', answer)
        contexte_match = re.search(r'CONTEXTE_AUTEUR_RUSSE:([^\n]+)', answer)
        
        if centralite_match:
            df_textes.loc[i, "Centralité de la littérature"] = int(centralite_match.group(1))
        
        if noms_match:
            noms_text = noms_match.group(1).strip()
            
            # Check if the response is EXACTLY "aucun" (case where there are truly no names)
            if noms_text.lower().strip() == 'aucun' or noms_text.strip() == '[aucun]':
                df_textes.loc[i, "Noms_propres_identifies"] = ""
            else:
                # Clean only individual elements that contain "aucun"
                # For example, "Grossman(aucun)" becomes "Grossman"
                noms_text = re.sub(r'\([^)]*aucun[^)]*\)', '', noms_text)
                noms_text = re.sub(r',\s*,', ',', noms_text)  # Nettoyer les virgules doubles
                noms_text = noms_text.strip(' ,/')
                
                # If nothing remains after cleaning, set to empty
                if not noms_text or noms_text.isspace():
                    df_textes.loc[i, "Noms_propres_identifies"] = ""
                else:
                    df_textes.loc[i, "Noms_propres_identifies"] = noms_text
            
        if contexte_match:
            df_textes.loc[i, "Contexte_citation_auteur_russe"] = contexte_match.group(1).strip()
                
    except Exception as e:
        print(f"❌ Erreur ligne {i+1} : {e}")

# Results
print("\n" + "="*70)

df_textes.to_csv("LLM_sample_chunk_CAIRN.csv", index=False)
print(f"\n Fichier sauvegardé sous : LLM_sample_chunk_CAIRN.csv")
print(f" Nombre de lignes sauvegardées : {len(df_textes)}")
```
