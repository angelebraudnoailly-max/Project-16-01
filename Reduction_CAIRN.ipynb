### The following code was used to reduce the size of my CAIRN csv in order to use it via render.

### 1. Library Import

```python
import pandas as pd
```

### 2. Data Import

```python
df = pd.read_csv("CAIRN_LDA_FINAL2.csv")
```

### 3. Data Manipulation

```python
# Colomns to keep
cols_to_keep = [
    "url",
    "Unnamed: 0",
    "titre",
    "auteur",
    "section",
    "twitter_card",
    "journal",
    "annee",
    "numero",
    "name",
    "any_author_in_chunk",
    "author_present_in_article_x",
    "author_present_in_article_y",
    "author_present_in_article",
    "langue",
    "tokens",
    "topics_proportions_concat",
    "top_2_topics_with_prop",
    "top_4_topics_with_prop"
]

df_light = df[cols_to_keep]

bool_cols = [
    "any_author_in_chunk",
    "author_present_in_article_x",
    "author_present_in_article_y",
    "author_present_in_article"
]

for col in bool_cols:
    df_light[col] = df_light[col].apply(lambda x: False if pd.isna(x) else True if str(x) in ["1", "1.0", "True", "true"] else False)

df_light.to_csv("CAIRN_LDA_LIGHT.csv", index=False, encoding="utf-8")
print("CSV allégé créé : CAIRN_LDA_LIGHT.csv")
```
