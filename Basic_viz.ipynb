```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('LeMonde_auteur_filtré50.csv')
df2=pd.read_csv('CAIRN_real_final.csv')
display(df, df2)
```

```python
# Méthode 1 : Directe (si la colonne s'appelle exactement 'année')
annee_min = df2['annee'].min()
annee_max = df2['annee'].max()
print(f"Année minimum : {annee_min}")
print(f"Année maximum : {annee_max}")
```

```python
# Convertir la colonne en datetime (ISO 8601)
df['date_published'] = pd.to_datetime(df['date_published'], errors='coerce')

# Vérifier les dates non converties
print("Dates non converties :", df['date_published'].isna().sum())

# Supprimer les lignes avec dates invalides
df = df.dropna(subset=['date_published'])

# Extraire l'année
df['année'] = df['date_published'].dt.year

# Compter le nombre d'articles par année
distribution = df['année'].value_counts().sort_index()

# Afficher la distribution
print(distribution)

# Visualiser
plt.figure(figsize=(14,6))
distribution.plot(kind='bar')
plt.xlabel('Year')
plt.ylabel('Number of articles')
plt.title('Distribution of articles by year in Le Monde dataset')
plt.show()
```
